```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

240 words at 22:35
linear extrapolate from there
word limit is 4k


```python
word_counts = np.array([240,313,386,404,432])
times_from_start = np.array([0,3,16,21,37])

def func(x,m,c):
    return x*m + c
```

```python
fit, cov = curve_fit(func,times_from_start,word_counts)


plt.plot(times_from_start,word_counts, "x", ms=10)
plt.plot(times_from_start, func(times_from_start, *fit), linestyle='dashed')
plt.grid()
plt.xlabel("Minutes from start")
plt.ylabel("Word Count")

#lazy error propogation but whynot
mean_end_time = (4000 - word_counts[0] - fit[1])/fit[0]
top_bound = (4000 - word_counts[0] - fit[1])/(fit[0] - np.sqrt(cov[0,0]))
bottom_bound = (4000 - word_counts[0] - fit[1])/(fit[0] + np.sqrt(cov[0,0]))


print("Estimated time remaining:", mean_end_time)
print("Upper Bound:", top_bound)
print("Lower Bound", bottom_bound)
```
