```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
word_counts = np.array([240,313,386,404,432,484,522,588,635,796,888,937,995,1262,1368,1800,2111,2168,2256,2440,2562,2620,2738,2944])
times_from_start = np.array([0,3,16,21,37,25+18,25+22,25+16,25+31,60,70,75,79,91,96,151,195,201,209,250,263,268,279,296])

def func(x,m,c):
    return x*m + c
```

```python
fit, cov = curve_fit(func,times_from_start,word_counts)

plt.plot(times_from_start,word_counts, "x", ms=10)
plt.plot(times_from_start, func(times_from_start, *fit), linestyle='dashed')
plt.grid()
plt.xlabel("Minutes from start")
plt.ylabel("Word Count")

#lazy error propogation but whynot
mean_end_time = (4000 - word_counts[-1] - fit[1])/fit[0]
top_bound = (4000 - word_counts[-1] - fit[1])/(fit[0] - np.sqrt(cov[0,0]))
bottom_bound = (4000 - word_counts[-1] - fit[1])/(fit[0] + np.sqrt(cov[0,0]))


print("Estimated time remaining:", mean_end_time)
print("Upper Bound:", top_bound)
print("Lower Bound", bottom_bound)
```
